### 1. **模型來源**

#### **This Waifu Does Not Exist**
* https://www.thiswaifudoesnotexist.net/

This Waifu Does Not Exist 是一個基於 GAN , 專門用來生成**不存在的二次元角色**。

這些二次元角色不是來自任何動畫、漫畫或遊戲, 而是利用 StyleGAN 模型根據學習到的數據，自動創造出的全新動漫人物。

進入此網站後會隨機生成一張全新的二次元角色圖片, 不滿意可以按底下的 REFRESH 刷新, 生成新的一張角色圖。

StyleGAN工作原理:
* 從大量二次元角色圖片中學習風格與特徵（如眼睛、髮型、表情等）。
* 生成器 (Generator) 嘗試創造一張新的角色圖片, 而判別器 (Discriminator) 負責檢測它是否逼真。
* 透過不斷對抗訓練，最終讓 AI 能夠創造出高品質的二次元角色，且幾乎無法區分真假。

### 4. **GAN v.s. Diffusion Models**
#### **Q. 為什麼現在較沒有人在使用GAN來生圖？**

##### 1. **Diffusion Models 生成的圖片質量更高且細節更豐富**
* GAN Models 會在背景或邊緣區域產生重複模式 (如多餘的眼睛、扭曲的手指等)
* Diffusion Models 更能保留細膩的紋理, 特別適合複雜的場景（如寫實風景、超解析度影像等）

##### 2. **Diffusion Models 訓練更穩定、容易控制**
* GAN Models 訓練時容易失衡，導致模型生成質量下降。需要大量調整參數設定
* Diffusion Models 訓練更穩定，不需要手動調整對抗平衡

##### 3. **Diffusion Models 較適用於文本生成圖像 (Text-to-Image)**
* GAN Models 較適合**圖像生成圖像 (Image-to-Image)** , 如風格複製、超分辨率等。但不適合文本輸入生圖
* Diffusion Models 適合**文本驅動的圖像生成**, 更易實現精緻的AI繪圖

##### 4. **Diffusion Models 多樣性與可控性較高**
* GAN Models 如果訓練數據有限, 可能會反覆生成類似的圖像，而不會做出太多創新
* Diffusion Models 可以通過**不同的調控方式**, 較精確的控制生成圖像內容

##### 5. **Diffusion Models 計算成本較低**
