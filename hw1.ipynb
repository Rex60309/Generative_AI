# 畫函數圖形 - **利薩茹曲線 (Lissajous Curve)**
## 簡介
李薩如曲線是一種 **參數方程** 繪製的圖形，它描述了兩個 **垂直振盪的正弦波** 之間的關係。  
它的方程式如下:

$$
x = \sin(a t + \delta)
$$
$$
y = \sin(b t)
$$

其中：
- **\($ a $)**: X 軸的頻率
- **\($ b $)**: Y 軸的頻率
- **\($ \delta $)**: 相位差（單位：弧度）
- **\($ t $)**: 時間變數，範圍 \($ 0 $) 到 \($ 2\pi $)

# 實作

* import標準套件

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

* define繪製利薩如曲線的function

```python
def plot_lissajous(a, b, delta):
    t = np.linspace(0, 2 * np.pi, 1000)
    x = np.sin(a * t + delta)
    y = np.sin(b * t)

    plt.figure(figsize=(6,6))
    plt.plot(x, y, color="blue", linewidth=2)
    plt.title(f"Lissajous Curve (a={a}, b={b}, δ={delta:.2f} rad)")
    plt.xlabel("X Axis")
    plt.ylabel("Y Axis")
    plt.axis("equal")
    plt.grid(True)
    plt.plot()
```

* 設定參數並呼叫function繪圖

```python
a = 3
b = 2
delta = np.pi / 2

plot_lissajous(a, b, delta)
```
